```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dataset/3_before.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
def get_mask(img, threshold = 230):
    ret, mask = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)
    return cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
```

```python
mask = get_mask(img)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB))
```

```python
dst = cv2.inpaint(img, mask, 20, cv2.INPAINT_TELEA)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
```

```python
cv2.imwrite('dataset/3_out.jpg', cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
```
