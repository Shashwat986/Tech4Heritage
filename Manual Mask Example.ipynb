```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
from skimage import io
```

```python
img = io.imread('https://github.com/Shashwat986/Tech4Heritage/blob/master/dataset/2_before.jpg?raw=true')
plt.imshow(img)
```

```python
mask = io.imread('https://github.com/Shashwat986/Tech4Heritage/blob/master/dataset/2_mask.jpg?raw=true')
mask = cv2.cvtColor(mask, cv2.COLOR_RGB2GRAY)
plt.imshow(mask)
```

```python
dst = cv2.inpaint(img, mask, 40, cv2.INPAINT_TELEA)
plt.imshow(dst)
```
