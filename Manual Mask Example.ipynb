```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dataset/2_before.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
mask = cv2.imread('dataset/2_mask.jpg', 0)
plt.imshow(cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB))
```

```python
dst = cv2.inpaint(img,mask,40,cv2.INPAINT_TELEA)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
```
